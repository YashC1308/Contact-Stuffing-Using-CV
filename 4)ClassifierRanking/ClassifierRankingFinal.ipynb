```python
import numpy as np
from matplotlib import pyplot as plt
from keras.models import load_model
from numpy import argmax
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from tensorflow.keras.datasets import mnist

# load and prepare the image
def load_image(filename):
 # load the image
 img = load_img(filename, color_mode = "grayscale", target_size=(28, 28))
 # convert to array
 img = img_to_array(img)
 # reshape into a single sample with 1 channel
 img = img.reshape(1, 28, 28, 1)
 # prepare pixel data
 img = img.astype('float32')
 img = img / 255.0
 return img
```

```python
import keras.backend as K


def euclidean_distance_loss(y_true, y_pred):
    """
    Euclidean distance loss
    https://en.wikipedia.org/wiki/Euclidean_distance
    :param y_true: TensorFlow/Theano tensor
    :param y_pred: TensorFlow/Theano tensor of the same shape as y_true
    :return: float
    """
    return K.sqrt(K.sum(K.square(y_pred - y_true), axis=-1))

ranking_list = []
#ranking_list = list(np.load("/content/drive/MyDrive/RankingEuclideanUnicCosine60000.npy",allow_pickle = True))
print("Data loaded : ",len(ranking_list)," datapoints loaded")

for i in range(0,len(unicode_data)):
  if(i%1000==0):
    print(i)
  if(i%10000 == 0):
    temp = np.array(ranking_list)
    np.save(f'../Assets/RankingEuclideanUnicCosine{i}.npy',temp)
    print(i)
    print("Success")

  output = np.zeros(100)

  #plt.imshow(x_train[i].reshape(28,28), cmap = "gray")
  #plt.show()


  lat = encoder.predict(unicode_data[i].reshape(-1,28,28),verbose=0)


  for j in range(100):
    output[j] = euclidean_distance_loss(lat, base_points[j])
    #output[j] = cosine_similarity(lat, base_points[j], axis=-1).numpy()[0]

  min_at_index = np.argmin(output)

  if(min_at_index>=0 and min_at_index<10):
    label = 0
  elif(min_at_index>=10 and min_at_index<20):
    label = 1
  elif(min_at_index>=20 and min_at_index<30):
    label = 2
  elif(min_at_index>=30 and min_at_index<40):
    label = 3
  elif(min_at_index>=40 and min_at_index<50):
    label = 4
  elif(min_at_index>=50 and min_at_index<60):
    label = 5
  elif(min_at_index>=60 and min_at_index<70):
    label = 6
  elif(min_at_index>=70 and min_at_index<80):
    label = 7
  elif(min_at_index>=80 and min_at_index<90):
    label = 8
  elif(min_at_index>=90 and min_at_index<100):
    label = 9

  ranking_list.append(np.array([unicode_data[i],output[min_at_index],label]))
  #print(output)
  #print(output[np.argmin(output)])
  #print(label)

ranking_list = np.array(ranking_list)
np.save('../Assets/RankingEuclidean_FinalRankedUnic.npy',ranking_list)
```

```python
ranking = np.load("../Assets/RankingEuclidean_FinalRankedUnic.npy",allow_pickle = True)

image_data = []
distance_data = []
label_data = []

for i in range(len(ranking)):
  image_data.append(ranking[i][0])
  distance_data.append(ranking[i][1])
  label_data.append(ranking[i][2])

image_data = np.array(image_data)
distance_data = np.array(distance_data)
label_data = np.array(label_data)
#print(distance_data)

ind = np.argsort(distance_data)
distance_data = distance_data[ind]
image_data = image_data[ind]
label_data = label_data[ind]

#print(distance_data)
#print(image_data)

#plt.imshow(image_data[2].reshape(28,28), cmap = "gray")
#plt.show()

```

```python
final_ranking = []
for i in range(len(image_data)):
  final_ranking.append(np.array([image_data[i],distance_data[i],label_data[i]]))

#print(final_ranking)
np.save('../Assets/RankingEuclidean_FinalRankedUnic.npy',final_ranking)
print("done")
```
